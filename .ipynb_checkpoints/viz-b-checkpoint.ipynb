## Stats

```python
mpg_data = pd.read_csv('mpg.csv')
```

```python
Four_cylinders_cars = mpg_data[mpg_data['cylinders'] == 4]
```

```python
mpg_by_origin = Four_cylinders_cars.groupby('origin')['mpg'].mean()
```

```python
mpg_by_origin.head(2)
```

## Conclusions
1. The fuel consumption is higher for larger number of cylinders
2. The fuel consumption is lower for modern cars (in general)
3. Japanease cars are the most efficient when it comes to fuel consumption
